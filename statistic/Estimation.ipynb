# Estimation

CI = sample mean +- ME  
ME = z_alpha_over_2 * SE  
SE = sigma / sqrt(n)   


CI = sample proportion +- ME  
ME = z_alpha_over_2 * SE  
SE = sqrt(p_hat * (1 - p_hat) / n)



Population variance -> population standard deviation  
Sample variance:
- biased sample variance -> uncorrected sample standard deviation (biased)
- unbiased sample variance -> corrected sample standard deviation (biased)
- ....................... -> Unbiased sample standard deviation

```python
import numpy as np
import scipy.stats as stats

results = {}

# 1. Confidence Interval for proportion
n, x = 500, 300
p_hat = x / n
cl = 0.95
alpha = 1 - cl
z_alpha_over_2 = stats.norm.ppf(1 - alpha / 2)
se = np.sqrt(p_hat * (1 - p_hat) / n)
ci = (p_hat - z_alpha_over_2 * se, p_hat + z_alpha_over_2 * se)
results["Q1"] = {"point_estimate": p_hat, "confidence interval": ci}

# 2. Confidence interval for mean
sample_mean = 22300
std_dev = 1200
n = 100
conf_levels = [0.90, 0.95, 0.99]
ci = {}
for cl in conf_levels:
    z_alpha_over_2 = stats.norm.ppf(1 - (1 - cl) / 2)
    ME = z_alpha_over_2 * std_dev / np.sqrt(n)
    ci[f"{int(cl * 100)}%"] = (sample_mean - ME, sample_mean + ME)
results["Q2"] = ci

# 3. Confidence interval for proportion
n, x = 700, 350
p_hat = x / n
cl = 0.9
alpha = 1 - cl
z_alpha_over_2 = stats.norm.ppf(1 - alpha / 2)
se = np.sqrt(p_hat * (1 - p_hat) / n)
ci = (p_hat - z_alpha_over_2 * se, p_hat + z_alpha_over_2 * se)
results["Q3"] = {"confidence_interval": ci}

# 4. Confidence Interval for proportion
n, x = 1050, 851
p_hat = x / n
cl = 0.95
alpha = 1 - cl
z_alpha_over_2 = stats.norm.ppf(1 - alpha / 2)
se = np.sqrt(p_hat * (1 - p_hat) / n)
ci = (p_hat - z_alpha_over_2 * se, p_hat + z_alpha_over_2 * se)
results["Q4"] = {"confidence_interval": ci}

# 5. Confidence Interval for mean
n, sample_mean, std_dev = 64, 48, 4
cl = 0.95
alpha = 1 - cl
z_alpha_over_2 = stats.norm.ppf(1 - alpha / 2)
me = z_alpha_over_2 * std_dev / np.sqrt(n)
ci = (sample_mean - me, sample_mean + me)
results["Q5"] = {"confidence_interval": ci}

# 6. Confidence Interval for mean
speeds = np.array(
    [
        49,
        56,
        83,
        65,
        68,
        60,
        76,
        86,
        74,
        53,
        71,
        65,
        74,
        72,
        64,
        42,
        62,
        62,
        58,
        82,
        78,
        55,
        65,
        87,
        56,
        50,
        71,
        58,
        57,
        75,
        58,
        64,
        86,
        56,
        45,
        73,
        54,
        86,
        70,
        73,
    ]
)

n = len(speeds)
sample_mean = np.mean(speeds)
std_dev = np.std(speeds, ddof=1)
cl = 0.95
alpha = 1 - cl
t_alpha_over_two = stats.t.ppf(1 - alpha / 2, n - 1)
me = t_alpha_over_two * std_dev / np.sqrt(n)
ci = (sample_mean - me, sample_mean + me)
results["Q6"] = ci

# 7. Confidence Interval for mean
tomatoes = np.array([96, 18, 94, 71, 56, 111, 144, 92, 135, 62, 71, 87, 31, 173, 100])
n = len(tomatoes)
sample_mean = np.mean(tomatoes)
std_dev = np.std(tomatoes, ddof=1)
cl = 0.9
alpha = 1 - cl
t_alpha_over_two = stats.t.ppf(1 - alpha / 2, n - 1)
me = t_alpha_over_two * std_dev / np.sqrt(n)
ci = (sample_mean - me, sample_mean + me)
results["Q7"] = {"confidence_interval1": ci}
# 7. Confidence Interval for proportion
tomatoes_less_than_80 = [t for t in tomatoes if t <= 80]
# print(tomatoes_less_than_80)
n = len(tomatoes)
x = len(tomatoes_less_than_80)
p_hat = x / n
print(f"p_hat = {p_hat}")
cl = 0.95
alpha = 1 - cl
z_alpha_over_two = stats.norm.ppf(1 - alpha / 2)
print(f"z_alpha_over_two = {z_alpha_over_two}")
me = z_alpha_over_two * np.sqrt(p_hat * (1 - p_hat) / n)
ci = (p_hat - me, p_hat - me)
results["Q7"] = {**results["Q7"], "confidence_interval2": ci}


results
```
